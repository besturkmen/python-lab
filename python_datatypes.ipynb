```python
print("Benim adım {}".format('Beste'))
print("Benim adım {}, yasim {}".format('Beste', 29))

print("Benim adım {0}, yasim {1}".format('Beste', 29))
print("Benim adım {1}, yasim {0}".format('Beste', 29))

print("Benim adım {ad}, yasim {yas}".format(ad='Beste', yas=29))
```
